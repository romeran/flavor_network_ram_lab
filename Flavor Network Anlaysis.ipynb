# Flavor Network Analysis

Loading the recipies data to a Dataframe, such that
* Every row is a recipe
* Every column is an ingredient

Each cell in this table is either 0 if the ingredient is not in the recipe or 1 if it’s in the recipe.

```python
import csv
import pandas as pd

recipies_file = './Desktop/srep00196-s3.csv'
```

```python
lst = []
for row in csv.reader(open(recipies_file, 'r'), delimiter=','):
    recipe = {ingredient: 1 for ingredient in row[1::]}
    recipe['cuisine'] = row[0]
    lst.append(recipe)
    
data = pd.DataFrame(lst).fillna(0)
```

## UMAP 
Sampling 20% of the data, and fitting a UMAP model

```python
import umap
import umap.plot
import sklearn

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

le.fit(data['cuisine'])
data['encoded_cuisine'] = le.transform(data['cuisine'])
encoded_data = data.drop('cuisine', axis=1)

sampled_data = encoded_data.sample(frac=0.2).reset_index(drop=True)
```

```python
mapper = umap.UMAP().fit(sampled_data)
```

```python
hover_data = pd.DataFrame({'index':sampled_data.index,
                           'label':sampled_data.encoded_cuisine})
hover_data['cuisine'] = hover_data.label.map(
    {
        0:'African',
        1:'East Asian',
        2:'Eastern European',
        3:'Latin American',
        4:'Middle Eastern',
        5:'North American',
        6:'Northern European',
        7:'South Asian',
        8:'Southeast Asian',
        9:'Southern European',
        10:'Western European',
    }
)

umap.plot.output_notebook()
```

### UMAP - Euclidean Metric

UMAP with the default distance metric (euclidean) looks rather good. Recipies are nicely clustered by cuisines

```python
p = umap.plot.interactive(mapper, labels=sampled_data['encoded_cuisine'], hover_data=hover_data, theme='fire', point_size=3)
umap.plot.show(p)
```

### UMAP - Cosine Metric

Clustering seems less succesfull. Most likely the recipie representation should be adapted

```python
cosine_mapper = umap.UMAP(metric='cosine').fit(sampled_data)
```

```python
cosine_p = umap.plot.interactive(cosine_mapper, labels=sampled_data['encoded_cuisine'], hover_data=hover_data, theme='fire', point_size=3)
umap.plot.show(cosine_p)
```

### UMAP on a Sparse matrix

```python
sampled_sparse_data = encoded_data.sample(frac=0.2).reset_index(drop=True)

sparse = scipy.sparse.csr_matrix(sampled_sparse_data.values)
sparse

sparse_mapper = umap.UMAP(random_state=42, low_memory=True).fit(sparse)
umap.plot.points(sparse_mapper, values=np.arange(11300),  theme='viridis')
```
